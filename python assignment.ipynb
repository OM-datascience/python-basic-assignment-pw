**Theory Questions**

1. What is Python, and why is it popular?

   - Python is a simple programming language that is easy to learn. It is popular because it is used in many areas like making websites and analyzing data.
2. What is an interpreter in Python?

   - An interpreter in Python is a tool that runs Python code step by step.
3. What are pre-defined keywords in Python?

   - Pre-defined keywords are words like if, else, and while that Python uses for special tasks.
4. Can keywords be used as variable names?

   - No, keywords cannot be used as variable names because they are already used by Python.
5.  What is mutability in Python?

   - Mutability means an object like a list can be changed after it is made.

6.  Why are lists mutable, but tuples are immutable?

   - Lists can be changed because they are flexible, but tuples cannot be changed because they are fixed.
7.  What is the difference between "==" and "is" operators in Python?

   - "==" checks if two values are the same, while "is" checks if they are the same object.

9. What is type casting in Python?

  - Type casting is changing one data type to another, like turning a number into a word.
10. What is the difference between implicit and explicit type casting?
    - Implicit type casting happens automatically, like adding 1 and 1.0, while explicit type casting is done by us, like using int().
11. How does the purpose of conditional statements work in Python?
    - Conditional statements like if decide which part of code to run based on a condition.
12. What is the difference between for and while loops?
   - A for loop goes through a list a set number of times, while a while loop runs until a condition is false.
13. Describe a scenario where a while loop is more suitable than a for loop.
   - A while loop is better when you don’t know how many times to repeat, like asking for input until the user types "stop".

```python
''' 6. Write a function to demonstrate the behavior of mutable and immutable arguments'''
def modify_arguments(num, num_list):
    # Try to change the immutable argument (int)
    num += 10

    # Try to change the mutable argument (list)
    num_list.append(10)

    print("Inside function - num:", num)
    print("Inside function - num_list:", num_list)

# Immutable argument (int)
x = 5

# Mutable argument (list)
y = [1, 2, 3]

# Call the function
modify_arguments(x, y)

# Print values after function call
print("Outside function - x:", x)         # Still 5 (unchanged)
print("Outside function - y:", y)         # Changed: [1, 2, 3, 10]
```

```python
#Write a program that performs basic arithmetic operations on two user-input numbers.
# Ask the user to enter two numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

# Handle division carefully to avoid division by zero
if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (cannot divide by zero)"

# Display the results
print("\n--- Results ---")
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

```python
#Write a Python program to convert user input from string to integer, float, and boolean types.
# Ask the user for input
user_input = input("Enter a value: ")

print("\n--- Type Conversions ---")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer:", int_value)
except ValueError:
    print("Cannot convert to Integer.")

# Convert to float
try:
    float_value = float(user_input)
    print("Float:", float_value)
except ValueError:
    print("Cannot convert to Float.")

# Convert to boolean
# Any non-empty string (except "false" or "0") will be considered True
bool_value = bool(user_input)
print("Boolean:", bool_value)
```

```python
#Write code to demonstrate type casting with list elements.
# Original list of strings (representing numbers)
str_list = ['10', '20', '30', '40.5', '50']

print("Original list (strings):", str_list)

# Convert all elements to integers (skipping non-integer strings)
int_list = []
for item in str_list:
    try:
        int_list.append(int(item))
    except ValueError:
        print(f"Cannot convert '{item}' to int.")

print("List after casting to integers:", int_list)

# Convert all elements to floats
float_list = [float(item) for item in str_list] # Corrected line
print("List after casting to floats:", float_list)
```

```python
# Write a Python program to find the sum of all even numbers between 1 and 50.
# Initialize sum variable
even_sum = 0

# Loop through numbers from 1 to 50
for num in range(1, 51):
    if num % 2 == 0:  # Check if the number is even
        even_sum += num

# Print the result
print("The sum of all even numbers between 1 and 50 is:", even_sum)
```

```python
#Write a program to reverse a string using a while loop.
# Get input string from the user
original_str = input("Enter a string: ")

# Initialize variables
reversed_str = ""
index = len(original_str) - 1  # Start from the last character

# Loop to build the reversed string
while index >= 0:
    reversed_str += original_str[index]
    index -= 1

# Display the reversed string
print("Reversed string:", reversed_str)
```

```python

#Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# Get number input from the user
num = int(input("Enter a non-negative integer: "))

# Initialize factorial result and counter
factorial = 1
counter = num

# Check for valid input
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    # Calculate factorial using while loop
    while counter > 1:
        factorial *= counter
        counter -= 1

    print(f"Factorial of {num} is {factorial}")
```
